```python
class TicTacToc(object):
    """
    TicTacToc Game
    __init__ 
        ._board := express visited the place or not
        ._reader := Displayed board
        ._val := marker if 1 then 'X' else 'O'
        ._n := n x n matrix
    
    start:
        1.while (two condition) to play, then the result was success or deprecated
        2. Re-game Option
    
    _insert_checker()
        1. received the input row, col index
        2. Check the followings
            - type of row, col integer or not
            - the range of row and col is not over limitation
            - the position already visited or not
        3. Allocate the correct row, col
    _display(): Show the current board
    _success_fail(): Diagnose either markers win or keep going to compete
    _winning_message: Display who is the winner
    _replay: Ask whether do you want one more round
    _clean(): To regame, clean the all record
    """
    def __init__(self, n = 3):
        self._board = [[0] * n for _ in range(n)]
        self._reader = [['-'] * n for _ in range(n)]
        self._val = 1
        self._n = n
        
    def start(self):
        print('Game Start!')
        round_game, fin_round = 0, self._n ** 2
        
        game_stop = False
        while not game_stop and round_game < fin_round:
            print('\n')
            print('X turn') if self._val == 1 else print('O turn')
            self._insert_checker() 
            self._display()
            
            round_game += 1
            self._val *= -1
            game_stop = self._success_fail()
                
        if game_stop: self._winning_message()
        else: print('The Game was deprecated')
        print('Game End')
            
        if self._replay(): 
            self._clean()
            self.start()
        
        
    def _insert_checker(self):
        n = self._n
        while True:
            try:
                row = int(input("Enter a row: "))
                col = int(input("Enter a col: "))
            except ValueError:
                print("Please insert integer between 0 and {}".format(n))
                continue
        
            if not (-1 < row < n and -1 < col < n):
                print("Please insert integer between 0 and {}".format(n))
                continue
            if self._board[row][col] != 0:
                print('Insert Another ({},{})'.format(row,col))
                continue
        
            self._board[row][col] = self._val
            self._reader[row][col] = 'X' if self._val == 1 else 'O'
            return
        
    def _display(self):
        for row in self._reader: print(row)
        
    def _success_fail(self):
        n = self._n
        val = self._val * -1
        #Row Val Checker
        board = self._board        
        right_diag = all(val == board[i][i] for i in range(n))
        left_diag = all(val == board[i][j] for i,j in zip(range(n), reversed(range(n))))
        if right_diag or left_diag:
            return True
        
        for r, rowList in enumerate(board):
            if all(val == row_val for row_val in rowList): return True
            if all(val == col_val for col_val in [board[i][r] for i in range(n)]): return True
            
        return False        
        
    def _winning_message(self):
        print('********************************')
        val = self._val * -1
        if val == 1:
            print('X won')
        else:
            print('O won')
        print('********************************')
            
    def _replay(self):
        x = 'None'
        while not(x == 'Y' or x == 'N'):
            x = input('Do you want to do one more round? Y or N : ')
            print('Give a correct Response')
        return True if x == 'Y' else False
    
    def _clean(self):
        n = self._n
        self._board = [[0] * n for _ in range(n)]
        self._reader = [['-'] * n for _ in range(n)]
        self._val = 1
        
```
